# Simple model of infection

Each individual infects $c$ new people each day.

$I_{n+1} = I_n + (cI_n)$

So, $I_{n+1} = \lambda I_n$, where $\lambda := 1 + c$ -- **growth rate**

and $I_0 = 1$

```julia
I_0 = 1
```

```julia
c = 0.01 # average no. of people that each individual infects on each day
```

```julia
λ = 1 + c # \lambda <TAB>
```

```julia
T = 10 # final time

I = zeros(T)
```

```julia
I[1] = I_0 # Could use OffsetArrays.jl -- enables arbitrary indexing

for n in 1:T-1
    I[n+1] = λ * I[n]
end
```

```julia
I
```

```julia
using Plots
```

```julia
plot(I, m=:o, label="I[n]", leg=:topleft)
```

```julia
function run_infection(T, I_0=1)
    I = zeros(T)
    I[1] = I_0
    
    for n in 1:T-1
        I[n+1] = λ * I[n]
    end
    
    return I
end
```

```julia
methods(run_infection)
```

```julia
run_infection(10)
```

**Always separate data generation from plotting!**

```julia
I = run_infection(100)
plot(I, m=:o)
```

```julia
using Interact
```

```julia
end_T = 1000
@manipulate for T in slider(1:end_T, value=1)
    I_result = run_infection(T)
    
    plot(I_result, m=:o)
    xlims!(0, end_T)
end
```

## Logistic growth

Exponential growth is unrealistic: assumes that there are always more people to infect -- but population is finite.

$I_{n+1} = I_n + cI_n$

Each individual will be in contact with a fraction $\alpha$ of the population. At each contact there will be a probability $p$ that you infect each person. The population has size $N$.

So $c = p \alpha N$

Original model:

$I_{n+1} = I_n + p \alpha N I_n$ -- good approximation when almost everybody still susceptible.

New model:

Can only infect uninfected (susceptible) people!

$I_{n+1} = I_n + p \alpha S_n I_n$

$I_{n+1} = I_n + [p \alpha (N - I_n)] I_n = f(I_n)$

$I_{n+1} = I_n + \beta(I_n, S_n)I_n = f(I_n)$

```julia
p = 0.02
α = 0.01
N = 1000

β(I, S) = p * α * S

function run_infection(T, I_0=1)
    
    I = zeros(T)
    I[1] = I_0
    
    for n in 1:T-1
        I[n+1] = I[n] + β(I[n], N - I[n]) * I[n]
    end
    
    return I
end
```

```julia
I = run_infection(20)
```

```julia
plot(I, m=:o)
```

```julia
plot(I, m=:o, yscale=:log10)
```

```julia
I = run_infection(100)
plot(I, m=:o, yscale=:log10)
```

```julia
p = 0.01
α = 0.1
N = 100
T_end = 100

I = run_infection(T_end)
plot(I, m=:o)
```

**Sigmoid** shape - S-shaped

```julia
plot(I, m=:o, yscale=:log10)
```

"Logistic-type growth"

**Note:** All models are wrong, but some models are useful.

## More realism: Heterogeneity of individuals or groups

Idea: instead of modelling population globally, model each individuals

"Patch model": Local patche where population is **well-mixed**

Network model: Links between nodes

**Exponential growth with heterogeneity**

$I_{n+1} = \lambda_n I_n$ -- growth rate changes in time

```julia
p = 0.02
α = 0.01
N = 1000

β(I, S) = p * α * S

function run_infection(c_average = 1.1, T=20, I_0=1)
    
    I = zeros(T)
    I[1] = I_0
    
    for n in 1:T-1
        c = c_average + 0.1*randn() # Add Gaussian noise
        I[n+1] = I[n] + c * I[n]
    end
    
    return I
end
```

# Differential equations

```julia
using DifferentialEquations, Plots
```

```julia
x0 = 1.0
k = 1.0

f(x, k, t) = k*x # x'(t) = f(x,k,t)  f(function, parameter, time)
tspan = (0.0, 3.0)

prob = ODEProblem(f, x0, tspan, k)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)

plot(sol, lw=3, label="solution to the linear ODE")
plot!(sol.t, t->x0*exp(k*t), lw=3, ls=:dash, label="true solution")
```

```julia
using ModelingToolkit, DifferentialEquations
```

```julia
@parameters t, k
@variables x(t)
@derivatives D'~t

eqs = [D(x) ~ -k*x]
sys = ODESystem(eqs)

x0 = [x => 1.0]
p = [k => 1.0]
tspan = (0.0, 3.0)
prob = ODEProblem(sys, x0, tspan, p)
sol = solve(prob, Tsit5())

plot(sol)
```

```julia
@parameters t   
@variables x(t), y(t), z(t)
@derivatives D'~t

eqs = [D(x) ~ -*x + *y,
       D(y) ~ -*y + *z,
       D(z) ~       -*z]
sys = ODESystem(eqs)

u0 = [x => 1.0,
      y => 2.0,
      z => 0.0]

p = [ => 0.1,
      => 0.2,
      => 0.1]

tspan = (0.0, 3.0)
prob = ODEProblem(sys, u0, tspan, p)
sol = solve(prob, Tsit5())

plot(sol)
```

## SIR model without vital dynamics

$$\frac{dS}{dt} = -\frac{\beta I S}{N},\\
  \frac{dI}{dt} = \frac{\beta I S}{N} - \gamma I,\\
  \frac{dR}{dt} = \gamma I $$
  
Looking at $\frac{dS}{dt}$, at $t = 0, S = N$. $\beta$ is the number of people each infected person infects.

Looking at $\frac{dI}{dt}$, I grows by the number of people who went from S to I, and $\gamma$ is the chance of recover.

```julia
@parameters t  
@variables S(t) I(t) R(t)
@derivatives D'~t
eqs = [D(S) ~ -*S*I,
       D(I) ~ *S*I - *I,
       D(R) ~ *I]
sys = ODESystem(eqs)
u0 = [S => 0.99,
      I => 0.01,
      R => 0.0]
p = [ => 0.1,
      => 0.01]
tspan = (0.0, 200.0)
prob = ODEProblem(sys, u0, tspan, p)
sol = solve(prob, Tsit5())

plot(sol)
```

## SEIR model with vital dynamics

Assume birth rate = death rate = $\mu$.

$$ \frac{dS}{dt} = \mu N - \frac{\beta I S}{N} - \mu S\\
   \frac{dE}{dt} = \frac{\beta I S}{N}- a E - \mu E\\
   \frac{dI}{dt} = a E -\gamma I - \mu I\\
   \frac{dR}{dt} = \gamma I - \mu R $$
   
For this model, the basic reproduction number is:

$$ R_0 = \frac{a}{\mu + a}\frac{\beta}{\mu + \gamma} $$

```julia
@parameters t    a
@variables S(t) E(t) I(t) R(t)
@derivatives D'~t
eqs = [D(S) ~  - *S*I - *S,
       D(E) ~ *S*I - (a + )*E,
       D(I) ~ a*E - ( + )*I,
       D(R) ~ *I - *R]
sys = ODESystem(eqs)
u0 = [S => 0.99,
      E => 0.0,
      I => 0.1,
      R => 0.0]
p = [ => 520/365,
      => 1/30,
      => 774835/(65640000*365),
     a => 1/60]
tspan = (0.0, 365.0)
prob = ODEProblem(sys, u0, tspan, p)
sol = solve(prob, Tsit5())

plot(sol)
```
